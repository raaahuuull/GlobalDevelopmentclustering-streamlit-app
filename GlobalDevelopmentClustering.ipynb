Week 2 (Model Building)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, SpectralClustering
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

```

```python
currency_cols = ["GDP", "Health Exp/Capita", "Tourism Inbound", "Tourism Outbound"]

for col in currency_cols:
    if col in df.columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace("$", "", regex=False)
            .str.replace(",", "", regex=False)
            .str.strip()
        )
        df[col] = pd.to_numeric(df[col], errors="coerce")

df[currency_cols].head()
```

```python
for col in df.columns:
    if df[col].dtype == object:
     if df[col].astype(str).str.contains('%').any():
            df[col] = df[col].astype(str).str.replace('%', '', regex=False).str.strip()
            df[col] = pd.to_numeric(df[col], errors="coerce")
df.head()
```

```python
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isna().sum()
```

```python
df_model = df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

if 'Country' in df_model.columns:
    le_country = LabelEncoder()
    df_model['Country'] = le_country.fit_transform(df_model['Country'])

df_model.head()
```

```python
X = df_model.select_dtypes(include=['float64', 'int64'])
```

```python
print("Features used for clustering:", list(X.columns))
print("Shape of X:", X.shape)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
```

```python
pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pca_df.head()
```

```python
def evaluate_and_plot(model_name, labels, X_scaled, pca_df):

    plot_df = pca_df.copy()
    plot_df['Cluster'] = labels


    unique_labels = np.unique(labels)
    if len(unique_labels) > 1 and (len(unique_labels) < len(labels)):

        if -1 in unique_labels:
            mask = labels != -1
            if mask.sum() > len(np.unique(labels[mask])):
                sil = silhouette_score(X_scaled[mask], labels[mask])
            else:
                sil = None
        else:
            sil = silhouette_score(X_scaled, labels)
    else:
        sil = None

    print(f"\n===== {model_name} =====")
    print("Cluster counts:")
    print(pd.Series(labels).value_counts())
    if sil is not None:
        print("Silhouette Score:", round(sil, 4))
    else:
        print("Silhouette Score: Not applicable")

    # Plot in PCA space
    plt.figure(figsize=(7, 5))
    sns.scatterplot(
        data=plot_df,
        x="PC1",
        y="PC2",
        hue="Cluster",
        s=60,
        palette="tab10"
    )
    plt.title(f"{model_name} Clusters (PCA 2D)")
    plt.legend(title="Cluster", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.show()
```

MODEL 1 â€“ KMeans

```python
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)
df_model['Cluster_KMeans'] = kmeans_labels
evaluate_and_plot("KMeans", kmeans_labels, X_scaled, pca_df)
```

MODEL 2 â€“ Agglomerative Clustering

```python
agg = AgglomerativeClustering(n_clusters=5, linkage="ward")
agg_labels = agg.fit_predict(X_scaled)
df_model['Cluster_Agglomerative'] = agg_labels
evaluate_and_plot("Agglomerative", agg_labels, X_scaled, pca_df)
```

MODEL 3 â€“ DBSCAN

```python
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
df_model['Cluster_DBSCAN'] = dbscan_labels
evaluate_and_plot("DBSCAN", dbscan_labels, X_scaled, pca_df)
```

MODEL 4 â€“ Gaussian Mixture Model (GMM)

```python
gmm = GaussianMixture(n_components=5, random_state=42)
gmm.fit(X_scaled)
gmm_labels = gmm.fit_predict(X_scaled)

df_model['Cluster_GMM'] = gmm_labels

evaluate_and_plot("GMM", gmm_labels, X_scaled, pca_df)
```

MODEL 5 â€“ Mean Shift

```python
from sklearn.cluster import MeanShift
ms = MeanShift()
ms_labels = ms.fit_predict(X_scaled)
df_model['Cluster_MeanShift'] = ms_labels
evaluate_and_plot("MeanShift", ms_labels, X_scaled, pca_df)
```
